```python
# 5x5 Tic Tac Toe game using GUI
# using tkinter

# importing of all necessary libraries
import random
from copy import deepcopy
import tkinter
from tkinter import messagebox
from functools import partial
from tkinter import *


# variable to decide the turn of  player
symbol = 0

# Creating an empty ttt
global ttt
ttt = [[" " for x in range(5)] for y in range(5)]


def win_fun(a, l): # Check whi=o is the winner
	return ((a[0][0] == l and a[0][1] == l and a[0][2] == l and a[0][3] == l and a[0][4] == l) or
			(a[1][0] == l and a[1][1] == l and a[1][2] == l and a[1][3] == l and a[1][4] == l) or
			(a[2][0] == l and a[2][1] == l and a[2][2] == l and a[2][3] == l and a[2][4] == l) or
			(a[3][0] == l and a[3][1] == l and a[3][2] == l and a[3][3] == l and a[3][4] == l) or
			(a[4][0] == l and a[4][1] == l and a[4][2] == l and a[4][3] == l and a[4][4] == l) or
			(a[0][0] == l and a[1][0] == l and a[2][0] == l and a[3][0] == l and a[4][0] == l) or
			(a[0][1] == l and a[1][1] == l and a[2][1] == l and a[3][1] == l and a[4][1] == l) or
			(a[0][2] == l and a[1][2] == l and a[2][2] == l and a[3][2] == l and a[4][2] == l) or
			(a[0][3] == l and a[1][3] == l and a[2][3] == l and a[3][3] == l and a[4][3] == l) or
			(a[0][4] == l and a[1][4] == l and a[2][4] == l and a[3][4] == l and a[4][4] == l) or
			(a[0][0] == l and a[1][1] == l and a[2][2] == l and a[3][3] == l and a[4][4] == l) or
			(a[0][4] == l and a[1][3] == l and a[2][2] == l and a[3][1] == l and a[4][0] == l))


def mark_button(v1, v2, gb, lbl1, lbl2): # Configure text on button
	global symbol
	if ttt[v1][v2] == ' ':
		if symbol % 2 == 0:
			lbl1.config(state=DISABLED)
			lbl2.config(state=ACTIVE)
			ttt[v1][v2] = "X"
		else:
			lbl2.config(state=DISABLED)
			lbl1.config(state=ACTIVE)
			ttt[v1][v2] = "O"
		symbol += 1
		button[v1][v2].config(text=ttt[v1][v2])
	if win_fun(ttt, "X"):
		gb.destroy()
		box = messagebox.showinfo("Winner", "Player 1 Won")
	elif win_fun(ttt, "O"):
		gb.destroy()
		box = messagebox.showinfo("Winner", "Player 2 Won")
	elif(check_full()):
		gb.destroy()
		box = messagebox.showinfo("Tie Game", "Game Tie")


def check_free(v1, v2): # Check whether button can be pushed or not
	return ttt[v1][v2] == " "

def check_full():# Check whether the ttt is filledup
	flag = True
	for v1 in ttt:
		if(v1.count(' ') > 0):
			flag = False
	return flag


def t_board_player(game_board, label1, label2): # Create the GUI (player vs player)
	global button
	button = []
	for v1 in range(5):
		m = 3+v1
		button.append(v1)
		button[v1] = []
		for v2 in range(5):
			n = v2
			button[v1].append(v2)
			get_t = partial(mark_button, v1, v2, game_board, label1, label2)
			button[v1][v2] = Button(
				game_board, bd=5, command=get_t, height=6, width=10)
			button[v1][v2].grid(row=m, column=n)
	game_board.mainloop()


def ai(): # Decide systems next possible move
	next_move = []
	for v1 in range(len(ttt)):
		for v2 in range(len(ttt[v1])):
			if ttt[v1][v2] == ' ':
				next_move.append([v1, v2])
	move = []
	if next_move == []:
		return
	else:
		for let in ['O', 'X']:
			for v1 in next_move:
				boardcopy = deepcopy(ttt)
				boardcopy[v1[0]][v1[1]] = let
				if win_fun(boardcopy, let):
					return v1
		corner = []
		for v1 in next_move:
			if v1 in [[0, 0], [0, 4], [4, 0], [4, 4]]:
				corner.append(v1)
		if len(corner) > 0:
			move = random.randint(0, len(corner)-1)
			return corner[move]
		edge = []
		for v1 in next_move:
			if v1 in [[0, 1], [0, 2], [0, 3], [1, 0], [1, 1], [1, 2], [1, 3], [1, 4], [2, 0], [2, 1], [2, 2], [2, 3], [2, 4], [3, 0], [3, 1], [3,2], [3, 3], [3, 4], [4, 1], [4, 2], [4, 3]]:
				edge.append(v1)
		if len(edge) > 0:
			move = random.randint(0, len(edge)-1)
			return edge[move]



def get_text_computer(v1, v2, gb, label1, label2): # Configure text on button (player vs system)
	global symbol
	if ttt[v1][v2] == ' ':
		if symbol % 2 == 0:
			label1.config(state=DISABLED)
			label2.config(state=ACTIVE)
			ttt[v1][v2] = "X"
		else:
			button[v1][v2].config(state=ACTIVE)
			label2.config(state=DISABLED)
			label1.config(state=ACTIVE)
			ttt[v1][v2] = "O"
		symbol += 1
		button[v1][v2].config(text=ttt[v1][v2])
	x = True
	if win_fun(ttt, "X"):
		gb.destroy()
		x = False
		box = messagebox.showinfo("Winner", "Player Won")
	elif win_fun(ttt, "O"):
		gb.destroy()
		x = False
		box = messagebox.showinfo("Winner", "Computer Won")
	elif(check_full()):
		gb.destroy()
		x = False
		box = messagebox.showinfo("Game Tie", "Game Tie")
	if(x):
		if symbol % 2 != 0:
			move = ai()
			button[move[0]][move[1]].config(state=DISABLED)
			get_text_computer(move[0], move[1], gb, label1, label2)


def t_board_computer(game_board, label1, label2): # Create the GUI (player vs system)
	global button
	button = []
	for v1 in range(5):
		m = 3+v1
		button.append(v1)
		button[v1] = []
		for v2 in range(5):
			n = v2
			button[v1].append(v2)
			get_t = partial(get_text_computer, v1, v2, game_board, label1, label2)
			button[v1][v2] = Button(
				game_board, bd=5, command=get_t, height=4, width=8)
			button[v1][v2].grid(row=m, column=n)
	game_board.mainloop()


def computer(game_board): # Initializeing the game ttt to play with system
	game_board.destroy()
	game_board = Tk()
	game_board.title("Tic Tac Toe")
	label1 = Button(game_board, text="Player : X", width=10)
	label1.grid(row=1, column=1)
	label2 = Button(game_board, text = "Computer : O",
				width = 10, state = DISABLED)
	
	label2.grid(row = 2, column = 1)
	t_board_computer(game_board, label1, label2)

def challenger(game_board): # Initialize the game ttt to play with another player
	game_board.destroy()
	game_board = Tk()
	game_board.title("Tic Tac Toe")
	label1 = Button(game_board, text = "Player 1 : X", width = 10)
	
	label1.grid(row = 1, column = 1)
	label2 = Button(game_board, text = "Player 2 : O",
				width = 10, state = DISABLED)
	
	label2.grid(row = 2, column = 1)
	t_board_player(game_board, label1, label2)

# main function
def start():
	m = Tk()
	m.geometry("250x250")
	m.title("Tic Tac Toe")
	pc = partial(computer, m)
	pl = partial(challenger, m)
	
	head = Button(m, text = "---Welcome to tic-tac-toe---",
				activeforeground = 'red',
				activebackground = "yellow", bg = "red",
				fg = "yellow", width = 500, font = 'summer', bd = 5)
	
	B1 = Button(m, text = "Single Player", command = pc,
				activeforeground = 'red',
				activebackground = "yellow", bg = "red",
				fg = "yellow", width = 500, font = 'summer', bd = 5)
	
	B2 = Button(m, text = "Multi Player", command = pl, activeforeground = 'red',
				activebackground = "yellow", bg = "red", fg = "yellow",
				width = 500, font = 'summer', bd = 5)
	
	B3 = Button(m, text = "Exit", command = m.destroy, activeforeground = 'red',
				activebackground = "yellow", bg = "red", fg = "yellow",
				width = 500, font = 'summer', bd = 5)
	head.pack(side = 'top')
	B1.pack(side = 'top')
	B2.pack(side = 'top')
	B3.pack(side = 'top')
	m.mainloop()

# Call main function
if __name__ == '__main__':
	start()

```
